```python
"""
test1 은 미래 3개월 후 S&P 수익률 +/- 을 예측 
test2 는 미래 3개월 후 S&P 수익률 +내에서 strong,moderate 구분을 위한 학습 
"""
test1= econ_train1
test2= econ_train2
```

### Numpy로 변환

```python
# numpy 로 변환

# test1 은 미래 3개월 후 S&P 수익률 +/- 을 예측 
test1.dropna(inplace=True)
X1 = np.array(test1.iloc[:,:-1]) 
y1 = np.array(test1.iloc[:,-1])

# test2 는 미래 3개월 후 S&P 수익률 +내에서 strong,moderate 구분을 위한 학습 
test2.dropna(inplace=True)
X2 = np.array(test2.iloc[:,:-1]) 
y2 = np.array(test2.iloc[:,-1])

# 트레이닝, 테스트 데이터 셋 나누기 
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.15, random_state=321)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.15, random_state=321)
```

## 학습관련 코드


```python
# 먼저 Test1 관련 학습 
from sklearn.tree import DecisionTreeRegressor
"""
model1은 미래 3개월 후 S&P 수익률 +/- 을 예측 
"""

# 학습코드 
model = DecisionTreeRegressor(criterion='mae', max_depth=70, max_features=None,
                      max_leaf_nodes=None, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      presort=False, random_state=None, splitter='random').fit(X1_train,y1_train)

# 테스트 
dt_pred =model.predict(X1_test)
print('결정 트리 정확도: {0:.4f}'.format(accuracy_score(y1_test, dt_pred)))
print(dt_pred)
```

```python
# Test2 관련 학습 
from sklearn.tree import DecisionTreeRegressor
"""
Model2 는 +내에서 strong, moderate 구분 
"""
# 학습
model2 = DecisionTreeRegressor(criterion='mae', max_depth=70, max_features=None,
                      max_leaf_nodes=None, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      presort=False, random_state=None, splitter='random').fit(X2_train,y2_train)
#테스트 
dt_pred =model2.predict(X2_test)
print('결정 트리 정확도: {0:.4f}'.format(accuracy_score(y2_test, dt_pred)))
print(dt_pred)
```
